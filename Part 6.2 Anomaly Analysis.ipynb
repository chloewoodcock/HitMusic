```python
import matplotlib.pyplot as plt
```

```python
Categories = ['Death', 'Viral Video', 'Superbowl', 'Film', 'TV performance', 'Other music', 'Advert', 'Viral Dance', 'Sleeper Hit']
Counts = [19, 5, 4, 3, 3, 2, 2, 2, 2]
```

```python
def addlabels(x,y,z):
    for i in range(len(x)):
        plt.text(i, y[i] - 1.5, z[i], ha = 'center', color="black")
```

```python
plt.figure(figsize=(7,3))
plt.bar(Categories, Counts, color = 'orange')
addlabels(Categories, Counts, Counts)
plt.ylabel("Count")
plt.xlabel("Reason")
plt.title("Reasons for late song peak")
plt.xticks(rotation=45)
```
