# Printing out Null Lyrics

```python
import pandas as pd
```

```python
df = pd.read_csv(r'2010_2020_noDuplicates_lyrics1.csv')
```

```python
df
```

```python
nans = []
song_nans = []
artist_nans = []
count = 0
for x in range(0, 5250, 1):
    if str('nan') == str(df['lyrics'][x]):
        count +=1
        nans.append(x)
        song_nans.append(df['song'][x])
        artist_nans.append(df['artist'][x])
```

```python
count
```

```python
d = {'x': nans, 'song': song_nans, 'artist':artist_nans}
df_nan = pd.DataFrame(data=d)
```
