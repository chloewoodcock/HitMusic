## Adding chart date back into dataset

```python
import pandas as pd
```

```python
df2_test = pd.read_csv(r'2010_2020_lyrics1_sentiment_spotify.csv')
```

```python
df_test = pd.read_csv(r'2010_2020_lyrics1_sentiment_spotify_release.csv')
```

```python
merged_df = df2_test.merge(df_test, how = 'inner', on = ['song', 'artist'])
```

```python
merged_df
```

```python
merged_df.info()
```

```python
date_ = merged_df['date']
release_date_ = merged_df['release_date1']
rank_ = merged_df['rank_y']
song_ = merged_df['song']
artist_ = merged_df['artist']
last_week_ = merged_df['last-week_y']
peak_rank_ = merged_df['peak-rank_y']
weeks_on_board_ = merged_df['weeks-on-board_y']
lyrics_ = merged_df['lyrics_y']
lyrics_scores_ = merged_df['lyrics_scores_y']
lyrics_neg_ = merged_df['lyrics_neg_y']
lyrics_neu_ = merged_df['lyrics_neu_y']
lyrics_pos_ = merged_df['lyrics_pos_y']
lyrics_compound_ = merged_df['lyrics_compound_y']
lyrics_label_ = merged_df['lyrics_label_y']
spotify_pop_ = merged_df['spotify_pop_y']
spotify_id_ = merged_df['spotify_id_y']
spotify_explicit_ = merged_df['spotify_explicit_y']
```

```python
d = {'peak_date': date_, 'release_date': release_date_, 'song': song_, 'artist': artist_, 'peak_rank': peak_rank_, 
    'current_rank': rank_, 'last_rank': last_week_, 'weeks_board':weeks_on_board_, 'lyrics': lyrics_, 
    'lyrics_score': lyrics_scores_, 'lyrics_neg':lyrics_neg_, 'lyrics_neu':lyrics_neu_, 'lyrics_pos':lyrics_pos_,
    'lyrics_compound':lyrics_compound_, 'lyrics_label':lyrics_label_, 'spotify_pop': spotify_pop_, 
     'spotify_id':spotify_id_, 'spotify_explicit': spotify_explicit_}
df = pd.DataFrame(data=d)
```

```python
df
```

```python
df.to_csv (r'2010_2020_preTweet.csv', header=True)
```
