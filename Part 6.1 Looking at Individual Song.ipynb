```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("charts.csv", error_bad_lines=False)
```

```python
df_radio = df[df.song == 'Radioactive']
df_radio = df_radio[df_radio.artist == 'Imagine Dragons']
```

```python
df_radio = df_radio.sort_values(by=['weeks-on-board'])
```

```python
df_radio
```

```python
df_radio[df_radio['peak-rank'] == 3]
```

```python
plt.scatter(df_radio['weeks-on-board'], df_radio['rank'], alpha = 0.3, s=50)
plt.scatter(43, 3, color = 'navy', s=50)
plt.gca().invert_yaxis()
plt.xlabel("Weeks on the board")
plt.ylabel("Rank")
plt.title("Radioactive - Imagine Dragons: rank history")
```

```python
titles = ['lyrics', 'pre-release', 'now playing', 'title']
sent = [0.9579, 0.9952, 0.9743, 0.0]
```

```python
def addlabels(x,y,z):
    for i in range(len(x)):
        plt.text(i, y[i]+ 0.015, z[i], ha = 'center', color="black")
plt.bar(titles, sent, color = 'skyblue')
addlabels(titles, sent, sent)
plt.ylim([0, 1.1])
plt.xlabel("Variable")
plt.ylabel("Compound sentiment score")
plt.title("Radioactive - Imagine Dragons: Sentiments")
```
