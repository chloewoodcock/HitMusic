```python
#filter out each year
import datetime as dt
df['date'] = pd.to_datetime(df['date'])

data_2015 = df[df['date'].dt.year == 2015]
data_2016 = df[df['date'].dt.year == 2016]
data_2017 = df[df['date'].dt.year == 2017]
data_2018 = df[df['date'].dt.year == 2018]
data_2019 = df[df['date'].dt.year == 2019]
data_2020 = df[df['date'].dt.year == 2020]
```

```python
#concatenate together
data_2015_2020 = pd.concat([data_2020, data_2019, data_2018, data_2017, data_2016, data_2015])
```

```python
data_2015_2020 = data_2015_2020.reset_index()
```

```python
data_2015_2020
```

```python
#drop duplicates
data_duplicates = data_2015_2020.drop_duplicates(subset=['song', 'artist'])
```

```python
data_duplicates
```

```python
#filter years
data_2010 = df[df['date'].dt.year == 2010]
data_2011 = df[df['date'].dt.year == 2011]
data_2012 = df[df['date'].dt.year == 2012]
data_2013 = df[df['date'].dt.year == 2013]
data_2014 = df[df['date'].dt.year == 2014]
```

```python
data_2010_2020 = data_2010_2020.reset_index()
```
