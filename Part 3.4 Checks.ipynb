## Get difference between dates

```python
def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)
```

```python
count = 0
diff_day = []
for x in range(0, 5071, 1):
    try:
        day = days_between(df['release_date'][x], df['peak_date'][x])
        diff_day.append(day)
        if day < 7:
            print (x, df['song'][x], df['artist'][x])
            count += 1
    except:
        diff_day.append(17000)
```
