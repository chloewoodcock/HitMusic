```python
from datetime import datetime
from datetime import timedelta
endX = []
for x in range(0, 5015, 1):
    start = releaseX[x]
    Begindatestring = start
    Begindate = datetime.strptime(Begindatestring, "%Y-%m-%d")
    Enddate = Begindate - timedelta(days=14)
    one = str(Enddate.year)
    two = Enddate.month   
    two = ("%02d" % two)
    two = str(two)
    three = Enddate.day    
    three = ("%02d" % three)
    three = str(three)
    end_date = (one + "-" + two + "-" + three)
    end_date = str(end_date)
    endX.append(end_date)
```

```python
def change_song(song):
    song1=song
    song1.split(" ")
    if '(' in song1:
        index = song1.index("(")
        song1 = song1[:index]
        song = ''.join(song1)
    if '/' in song1:
        index = song1.index("/")
        song1 = song1[:index]
        song = ''.join(song1)
    elif '&' in song1:
        song1 = song1.replace("&", "")
        song = ''.join(song1)
    elif 'And' in song1:
        song1 = song1.replace(" And", " ")
        song = ''.join(song1)
    elif 'and' in song1:
        song1 = song1.replace(" and", " ")
        song = ''.join(song1)
    elif 'or' in song1:
        song1 = song1.replace(" or", " ")
        song = ''.join(song1)
    elif 'Not' in song1:
        song1 = song1.replace(" Not", " ")
        song = ''.join(song1)
    elif ',' in song1:
        song1 = song1.replace(",", "")
        song = ''.join(song1)
    elif '+' in song1:
        song1 = song1.replace("+", " ")
        song = ''.join(song1)
    elif '.' in song1:
        song1 = song1.replace(".", " ")
        song = ''.join(song1)
    elif '-' in song1:
        song1 = song1.replace("-", " ")
        song = ''.join(song1)
    elif '\'' in song1:
        song1 = song1.replace("\'", "")
        song = ''.join(song1)
    return song
```

```python
def change_artist(artist):
    artist1=artist
    artist1.split(" ")
    if 'Featuring' in artist1:
        index = artist1.index("Featuring")
        artist1 = artist1[:index]
        artist = ''.join(artist1)
    elif '&' in artist1:
        index = artist1.index("&")         
        artist1 = artist1[:index]
        artist = ''.join(artist1)
    elif '+' in artist1:
        index = artist1.index("+")
        artist1 = artist1[:index]
        artist = ''.join(artist1)
    elif ' And' in artist1:
        index = artist1.index(" And")
        artist1 = artist1[:index]
        artist = ''.join(artist1)
    elif ',' in artist1:
        index = artist1.index(",")
        artist1 = artist1[:index]
        artist = ''.join(artist1)  
    elif '(' in artist1:
        index = artist1.index("(")
        artist1 = artist1[:index]
        artist = ''.join(artist1)  
    elif ' With' in artist1:
        index = artist1.index(" With")
        artist1 = artist1[:index]
        artist = ''.join(artist1)
    elif ' X ' in artist1:
        artist1 = artist1.replace(" X ", "")
        artist = ''.join(artist1)
    elif ' x ' in artist1:
        artist1 = artist1.replace(" X ", "")
        artist = ''.join(artist1)
    return artist
```

```python
def create_headers(bearer_token):
    headers = {"Authorization": "Bearer {}".format(bearer_token)}
    return headers
```

```python
def create_url(keyword, start_date, end_date, max_results = 10):
    
    search_url = "https://api.twitter.com/2/tweets/search/all" #Change to the endpoint you want to collect data from

    #change params based on the endpoint you are using
    query_params = {'query': keyword,
                    'start_time': start_date,
                    'end_time': end_date,
                    'max_results': max_results,
                    'expansions': 'author_id,in_reply_to_user_id,geo.place_id',
                    'tweet.fields': 'id,text,author_id,in_reply_to_user_id,geo,conversation_id,created_at,lang,public_metrics,referenced_tweets,reply_settings,source',
                    'user.fields': 'id,name,username,created_at,description,public_metrics,verified',
                    'place.fields': 'full_name,id,country,country_code,geo,name,place_type',
                    'next_token': {}}
    return (search_url, query_params)
```

```python
def connect_to_endpoint(url, headers, params, next_token = None):
    params['next_token'] = next_token   #params object received from create_url function
    response = requests.request("GET", url, headers = headers, params = params)
    print("Endpoint Response Code: " + str(response.status_code))
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()
```
